```python
class weather_model(BaseModel):
    city:str
    date:str
```

```python
def get_weather(city):
    """
    get the weather of given city 
    """
    return f"the wather od {city}"
```

```python
from langchain_core.tools import StructuredTool

weather_tool=StructuredTool.from_function(
    func= get_weather,
    name="get_weather",
    description="get the real time weather",
    args_schema=weather_model

)
```
